```javascript
words_en= 'https://raw.githubusercontent.com/words/an-array-of-english-words/master/words.json';
words_es= 'https://raw.githubusercontent.com/words/an-array-of-spanish-words/master/palabras.json';

https.get(words_en, (resp) => {
  let data = '';

  // Traer cada parte
  resp.on('data', (chunk) => {
    data += chunk;
  });

  // Mostrar resultados
  resp.on('end', () => {
    console.log("data:",JSON.parse(data));
  });

}).on("error", (err) => {
  console.log("Error: " + err.message);
});
```
